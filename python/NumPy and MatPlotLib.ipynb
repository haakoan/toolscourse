### Array

We build N-dimensional arrays using **np.array**. Note that lists and arrays are very similar. However, they are completely different data structures.

````python
## 1D array
vec_1d = np.array([1,6,5,7])
## 2D array
vec_2d = np.array([[6,5,7], [0,1,3]])
````

````python
## access items
print(vec_1d[0]) #first element
print(vec_2d[1,2]) #second row, third column
````

Some **attributes** of the np.array class can be used to obtain information on the structure of a specific array.

````python
## check the shape of the array
print('shape of vec_1d:',vec_1d.shape)
print('shape of vec_2d:',vec_2d.shape)

## check the dimension
print('Ndim of vec_2d:',vec_2d.ndim)

## check the type
print('Type of vec_1d:',vec_1d.dtype)
````

#### Build arrays with built-in numpy functions

````python
## array filled with zeros
zeros_1D = np.zeros(10)  ## array of 10 elements
print(zeros_1D)
zeros_2D = np.zeros([2,5])  ## array of 2 rows and 5 columns
print(zeros_2D)

## ones
ones_1D = np.ones(10)
## note that it is straightforward to construct an array filled with any number
fives_1D = 5*np.ones(10)
print(fives_1D)

##arange
arange_vec = np.arange(0,5,0.2) ## np.arange(start, end, step)
print(arange_vec)
linsp_vec = np.linspace(1,5,10) ## np.linspace(start,end,# of elements)
print(linsp_vec)
````

#### Operations with arrays

Unlike lists, numpy allows operations on arrays without the use of loops.

````python
## sum of arrays 
vec1D_1 = np.array([1,6,5,7])
vec1D_2 = np.array([3,9,8,4])
sum_1D = vec1D_1 + vec1D_2
diff_1D = vec1D_1 - vec1D_2
print(sum_1D)
print(diff_1D)
````

**Note:** sum operations between two arrays with different dimensions follow algebraic rules. 

````python
## 1D array
vec_1d = np.array([1,6,5,7])
## 2D array
vec_2d = np.array([[1,6,5,7], [3,0,1,3]])
sum_multiD = vec_1d + vec_2d
print(sum_multiD)
````

This operation is allowed beacuse dim(vec_1d) = 4 and dim(vec_2d) = (2,4)

````python
## 1D array
vec_1d = np.array([1,5,7])
## 2D array
vec_2d = np.array([[1,6,5,7], [3,0,1,3]])
sum_multiD = vec_1d + vec_2d
print(sum_multiD)
````

On the other hand, this operation is not allowed

**Multiply and divide an array by a scalar**

````python
vec1 = np.array([[6,5,7], [1,0,4]])
vec2 = np.array([[1,6,5,7], [3,0,1,3], [0,1,3,4]])

k = 3
mvec1 = k*vec1
mvec2 = k*vec2

print(mvec1)
print(mvec2)
````

**Matrix multiplication**

````python
## 1D array
vec_1d = np.array([1,6,5,7])
## 2D array
vec_2d = np.array([[1,6,5,7], [3,0,1,3]])
multiD = vec_1d * vec_2d
print(multiD)
````

````python
vec1 = np.array([[6,5,7], [1,0,4]])
vec2 = np.array([[1,6,5,7], [3,0,1,3], [0,1,3,4]])
multiD = vec1 @ vec2 ## this operator support matrix multiplication
print(multiD)
````

Let's come back to the single subplot.

````python
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize=(7,7)) ##for other parameters (e.g., width_ratios) you can check the documentation
````

There is only one ax, so we don't need to access the array as before.

Before plotting them, we have to generate the data. 

We can define a function to calculate the cos of a given array.

````python
def calc_cos(x): ## argument is an array (or single variable)
    
    return np.cos(x)  ## built-in NumPy function that take the input x and return its cosin        
````

````python
## let's define the array xx as a linspace between -1 and 1

xx = np.linspace(-2*np.pi,2*np.pi,50)  ##pi is an object of NumPy
yy = calc_cos(xx) ## we assign the output of the function to the variable yy
````

Now we plot yy vs xx as a line. We can label the data in order to plot a legend.

````python
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize=(7,7)) ##for other parameters (e.g., width_ratios) you can check the documentation

## we can decide the color of the line (https://matplotlib.org/stable/gallery/color/named_colors.html)
c = 'Orangered'
## we can set the width of the line
lw = 2

## we can set a marker
marker = 'o'
## and assign another color
markerfacecolor = 'brown'

ax.plot(xx,yy, label = 'Cos', c = c, linewidth = lw, marker = marker, markerfacecolor = markerfacecolor,markeredgecolor = markerfacecolor)

plt.legend() #show the legend
plt.show()
````

We can plot another curve on top, let's plot $\sin^{2} + \cos^2$ vs xx and check that this is a costant line at 1

````python
def cos2sin2(x):
    
    sin2 = (np.sin(x)**2)
    cos2 = (np.cos(x)**2)
    return sin2 + cos2
````

````python
summ = cos2sin2(xx)

fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize=(7,7)) ##for other parameters (e.g., width_ratios) you can check the documentation

## we can decide the color of the line (https://matplotlib.org/stable/gallery/color/named_colors.html)
c = 'Orangered'
## we can set the width of the line
lw = 2

## we can set a marker
marker = 'o'
## and assign another color
markerfacecolor = 'brown'

ax.plot(xx,yy, label = 'Cos', c = c, linewidth = lw, marker = marker, markerfacecolor = markerfacecolor,markeredgecolor = markerfacecolor)


## now we can add the new line
## note that we can directly assign the parameters in the plot()

## we can also add a transaparency using alpha (it ranges from 0 to 1)
ax.plot(xx, summ, label = 'Cos^2 + Sin^2', linewidth = 2, color = 'darkblue', alpha = 0.3)




plt.legend()  ## show the legend 
plt.show()

````

### Scatter plot

In some cases we might be interested in a scatter plot rather than a line plot. 
Let's use the cosin again

````python
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize=(7,7)) ##for other parameters (e.g., width_ratios) you can check the documentation


c = 'Green'


ax.scatter(xx,yy, label = 'Cos', c = c) 

## let's plot np.sin(xx) in line, with another marker
ax.scatter(xx,np.sin(xx), label = 'Sin', c = 'cyan', marker = '^') 


plt.legend() #show the legend
plt.show()
````

We can also change the limits of the x axis and y axis and set labels

````python
#let's work with two subplots, the second will be a zoomed in view of the first one

fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize=(10,7)) ##for other parameters (e.g., width_ratios) you can check the documentation



ax[0].plot(xx,yy, label = 'Cos', c = c) 
ax[0].plot(xx,np.sin(xx), label = 'Sin', c = 'cyan') 

ax[1].plot(xx,yy, label = 'Cos', c = c) 
ax[1].plot(xx,np.sin(xx), label = 'Sin', c = 'cyan') 

## to zoom in the ax[1], we need to change the axes limits

ax[1].set_xlim(0,np.pi) ## limits of the x axis
ax[1].set_ylim(0,1) ## limits of the y axis

## set labels for the axes
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[1].set_xlabel('x')
#ax[0].set_ylabel('y') 


ax[0].legend() #show the legend in the first subplot
ax[1].legend() #show the legend in the second subplot
plt.show()
````
