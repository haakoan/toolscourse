```python
import numpy as np
import matplotlib.pyplot as plt

import numpy.random as rd
```

```python
x = np.linspace(-20,20,150)
```

```python
def damped_sin(x):
    return np.sin(x)*(np.exp(-(x**2)/30))
def damped_cos(x):
    return np.cos(x)*(np.exp(-(x**2)/30))
def noise(length):
    return rd.normal(0,0.2, length)
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (12,12))

y1 = damped_sin(x)
y2 = damped_cos(x)
y3 = noise(len(x))

## top left plot
ax[0,0].plot(x, y1, c = 'green', marker = 'o', markerfacecolor = 'None', markeredgecolor = 'darkgreen', alpha = 0.6, label = 'Damped sin wave')

## top right plot
ax[0,1].plot(x, y2, linestyle = '--',c = 'magenta', marker = 's', markerfacecolor = 'None', markeredgecolor = 'darkmagenta', alpha = 0.6, label = 'Damped cos wave')

## bottom left plot
ax[1,0].scatter(x, y3, c = 'grey', alpha = 0.6, label = 'Random Noise')

## bottom right plot

ax[1,1].plot(x,y1+y3, c = 'cyan', marker = 'o', markerfacecolor = 'None', markeredgecolor = 'darkcyan', alpha = 0.6, label = 'Noisy sin wave' )

ax[0,0].set_xlabel('x')
ax[0,0].set_ylabel('y1')
ax[0,1].set_xlabel('x')
ax[0,1].set_ylabel('y2')
ax[1,0].set_xlabel('x')
ax[1,0].set_ylabel('y3')
ax[1,1].set_xlabel('x')
ax[1,1].set_ylabel('y4')



ax[0,0].legend()
ax[0,1].legend()
ax[1,0].legend()
ax[1,1].legend()

plt.show()
```
